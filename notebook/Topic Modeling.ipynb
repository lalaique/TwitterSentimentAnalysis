### LDA with Gensim

#### Get data into correct format

```python
import pandas as pd
import pickle
from gensim.corpora import Dictionary
from gensim.models import Phrases # Compute bigrams.
import gensim

print("start")

# Read in dataset
df = pd.read_csv('dataset.csv', encoding ='utf_8')

# Desired doc format = ["hello", "hillary"]
docs = []

# limit = 1000
for index, row in df.iterrows():
    # uncomment if you want to add all words
    docs.append(row['final'].split(' '))
    # uncomment if you want to add only tags
#     tags_list = [i for i in row['final'].split() if i.startswith("#")]
#     if len(tags_list) > 0: 
#         docs.append(tags_list)
    # uncomment if you want to limit data size
#     if index > limit:
#         break
            
print("end")
```

#### Get corpus

```python
print("start")

# Create a dictionary representation of the documents
dictionary = Dictionary(docs)

# Filter out words that occur less than 20 documents, or more than 80% of the documents
below_nr=100
above_nr=0.6
dictionary.filter_extremes(no_below=below_nr, no_above=above_nr)

# Bag-of-words representation of the documents
corpus = [dictionary.doc2bow(doc) for doc in docs]

print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
print('no_below=', below_nr)
print('no_above=', above_nr)

NUM_TOPICS = 6
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=1)
ldamodel.save('model5.gensim')
topics = ldamodel.print_topics(num_words=10)
for topic in topics:
    print(topic)
    
print("end")
```

#### Display

```python
import pyLDAvis.gensim

lda_display = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```
